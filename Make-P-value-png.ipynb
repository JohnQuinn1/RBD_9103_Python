```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[10,6]
plt.rcParams['font.size']=16
from scipy.stats import chi2
plt.xkcd();
```

```python
x=np.linspace(0,50,200)
i=9
plt.plot(x, chi2.pdf(x,i+1), linewidth=2, label=f"DoF={i+1:}");
plt.legend();
plt.xlim([0,30])
plt.ylim([0,0.11])
plt.title(r' $\chi^2$ PDF');
plt.xlabel(r'$\chi^2$')
plt.ylabel('Probability');
xx=np.linspace(20,50,100)
yy=chi2.pdf(xx,i+1)
plt.fill_between(xx,yy,color='r')
plt.plot((22,24),(0.005,0.04),'r-',linewidth=2.0)
plt.text(22,0.045,"P-value",color='r',size=16);
plt.savefig("p-value.png")
```
